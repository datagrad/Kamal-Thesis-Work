```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_percentage_error
from google.colab import files  # Importing files module from google.colab

```

```python

# Load the dataset from the file
url = "https://raw.githubusercontent.com/datagrad/MS_Reference_MS-DS-LJMU-C13/main/city_day.csv"
df = pd.read_csv(url)

```

```python

# Filter data for specific cities
# cities_to_keep = ['Delhi']
cities_to_keep = ['Delhi', 'Bengaluru', 'Hyderabad']
df = df[df['City'].isin(cities_to_keep)]

# Retain only the desired columns: 'City', 'Date', and 'PM2.5'
df = df[['City', 'Date', 'PM2.5']]

# Convert 'Date' column to datetime data type
df['Date'] = pd.to_datetime(df['Date'])

# Ensure the data is sorted by Date
df = df.sort_values(by='Date')


# Replace NaN values in the 'PM2.5' column with next day's value (forward fill) for each city
df['PM2.5'] = df.groupby('City')['PM2.5'].fillna(method='ffill')

# Sort the DataFrame by the 'Date' column in ascending order
df.sort_values(by='Date', inplace=True)


# Creating lagged variables (e.g., t-1, t-2, t-3) to use as features
num_lags = 3  # using 3 lags as example
for lag in range(1, num_lags + 1):
    df[f'PM2.5_lag_{lag}'] = df.groupby('City')['PM2.5'].shift(lag)


# Interaction Terms Example: Lag1 * Month
# Extract Month
df['Month'] = df['Date'].dt.month
# Interaction Term
df['Lag1_Month'] = df['PM2.5_lag_1'] * df['Month']

# Data Transformation Example: Log Transformation
df['PM2.5_log'] = np.log1p(df['PM2.5'])

# Drop NaN values generated due to lags
df = df.dropna()

# Drop NaN rows created due to lagging
df = df.dropna()

# Perform a time-based train-test split
split_date = '2020-01-01'
train_df = df[df['Date'] < split_date]
test_df = df[df['Date'] >= split_date]

# Verify the datasets
train_df.head(), test_df.head()



# Save the DataFrame to a CSV file
filename = 'train_df.csv'
train_df.to_csv(filename, index=False)

filename_1 = 'test_df.csv'
test_df.to_csv(filename_1, index=False)

# Download the file to your local drive
files.download(filename)
files.download(filename_1)
```

Check for Stationarity:
You might want to check if your time series is stationary, using Augmented Dickey-Fuller Test, and differentiate it if necessary.

```python
from statsmodels.tsa.stattools import adfuller

# Performing Augmented Dickey-Fuller test on the PM2.5 time series for each city
for city in train_df['City'].unique():
    result = adfuller(train_df[train_df['City'] == city]['PM2.5'])
    print(f'ADF Statistic for {city}: {result[0]}')
    print(f'p-value: {result[1]}')
    print('Critical Values:', result[4])
    print('---')
```

The Augmented Dickey-Fuller (ADF) test is used to determine whether a unit root is present in an autoregressive model, which in turn helps to understand if the time series is stationary. The null hypothesis of the ADF test is that there is a unit root (i.e., the time series is not stationary), and the alternative hypothesis is that there is no unit root (i.e., the time series is stationary).

Your ADF Test Results:
Delhi: p-value = 0.0053 (< 0.05) --> Reject the null hypothesis --> Stationary
Bengaluru: p-value = 0.0000026 (< 0.05) --> Reject the null hypothesis --> Stationary
Hyderabad: p-value = 0.0065 (< 0.05) --> Reject the null hypothesis --> Stationary
All three time series seem to be stationary since the p-values are less than 0.05, allowing us to reject the null hypothesis that there is a unit root. Thus, you might not need differencing to make them stationary.

### Grid Search:
Systematically work through various combinations of orders and fit SARIMA models to observe which combination yields the best performance on your validation set.

```python
import itertools
import warnings

# Define the p, d, and q parameters to take any value between 0 and 2
p = d = q = range(0, 2)

# Generate all different combinations of p, d, and q triplets
pdq = list(itertools.product(p, d, q))

# Generate all different combinations of seasonal p, d, and q triplets
seasonal_pdq = [(x[0], x[1], x[2], 7) for x in list(itertools.product(p, d, q))]

warnings.filterwarnings("ignore") # specify to ignore warning messages

best_aic = float("inf")
best_pdq = None
best_seasonal_pdq = None

# Specify to ignore warning messages
warnings.filterwarnings("ignore")

for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            mod = SARIMAX(train_df['PM2.5'],
                          order=param,
                          seasonal_order=param_seasonal,
                          enforce_stationarity=False,
                          enforce_invertibility=False)
            results = mod.fit()
            # print('ARIMA{}x{} - AIC:{}'.format(param, param_seasonal, results.aic))
            if results.aic < best_aic:
                best_aic = results.aic
                best_pdq = param
                best_seasonal_pdq = param_seasonal
        except:
            continue

print('Best SARIMA model: {}x{} - AIC:{}'.format(best_pdq, best_seasonal_pdq, best_aic))
```

#### 3.5 Model Evaluation

##### 3.5.1 Introduction

Model evaluation is a critical stage in the predictive modeling process. It provides a quantifiable measure of the model’s predictive performance, offering insights into its accuracy, reliability, and applicability to real-world scenarios. In this research, several metrics were utilized to thoroughly evaluate the developed models' performance in predicting the Air Quality Index (AQI). A meticulous evaluation ensures that the selected models not only fit the training data well but also possess the capacity to generalize to unseen data, thereby ensuring their practical utility in real-world forecasting applications.

##### 3.5.2 Evaluation Metrics

- **Mean Absolute Percentage Error (MAPE):**
    - **Definition:** MAPE calculates the average of the absolute percentage errors between the actual and predicted values. It expresses the forecast errors as a percentage, which provides a scale-independent measure of prediction accuracy.
    - **Formula:**

    \[
    \text{MAPE} = \left( \frac{1}{n} \sum_{i=1}^{n} \left| \frac{A_i - F_i}{A_i} \right| \right) \times 100
    \]
    where \(A_i\) and \(F_i\) are the actual and forecasted values, respectively, and \(n\) is the number of observations.

    - **Application:** MAPE was employed to provide a clear and interpretable measure of the models' forecasting accuracy in percentage terms, facilitating easy comprehension and comparison of predictive performance.
    
- **Root Mean Squared Error (RMSE):**
    - **Definition:** RMSE calculates the square root of the average squared differences between the predicted and actual values. It provides a measure of the model’s predictive accuracy by quantifying the magnitude of the forecasting errors.
    - **Formula:**
    \[
    \text{RMSE} = \sqrt{\frac{1}{n}\sum_{i=1}^{n}(A_i - F_i)^2}
    \]
    where \(A_i\) and \(F_i\) represent the actual and forecasted values, respectively, and \(n\) is the number of observations.
    - **Application:** RMSE was utilized to assess the average magnitude of the models’ predictive errors, with a particular emphasis on penalizing larger errors, thereby ensuring that the models are not producing sporadic large forecasting discrepancies.
    
##### 7.5.3 Model Evaluation Process

- **In-Sample Evaluation:**
    - The models were first evaluated on the training data to assess their in-sample predictive performance, offering insights into how well the models capture the historical patterns and trends in the AQI data.

- **Out-of-Sample Evaluation:**
    - Subsequent to the in-sample evaluation, the models were assessed on the test data, providing a measure of their out-of-sample forecasting accuracy and generalization capability.
    
- **Residual Analysis:**
    - Residual plots and autocorrelation functions of the residuals were analyzed to ensure that the models adequately captured the underlying data structures and that the residuals resembled white noise, indicative of a well-specified model.
    
- **Model Comparison:**
    - The models’ MAPE and RMSE values were compared to gauge their relative predictive accuracies and select the model that offered the most reliable and accurate forecasts.

##### 3.5.4 Conclusion of Model Evaluation

The meticulous evaluation of the developed models, utilizing robust and interpretable metrics, provided a comprehensive understanding of their predictive capabilities. The use of both MAPE and RMSE ensured a thorough assessment of the models, considering both the relative and absolute magnitudes of their forecasting errors. This evaluation phase not only validated the models’ predictive accuracy but also their reliability and generalization capabilities, thereby affirming their applicability and utility in practical, real-world AQI forecasting scenarios. This methodical approach towards model evaluation underscores the reliability and credibility of the forecasting models developed in this research, ensuring they are robust, accurate, and applicable in practical scenarios.
